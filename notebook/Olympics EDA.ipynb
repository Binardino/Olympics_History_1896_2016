```python
#sorting countries per total amount of Medals
df_medal_per_country = df_athlete_events.groupby(['Team', 'Medal']).agg({'Medal':'count'})

df_medal_per_country.columns=['medal_count']

df_medal_per_country.sort_values(by=['medal_count'], ascending=False).head(50)
```

```python
#split athlete amount per sex
males_athlete = df_athlete_events.loc[df_athlete_events['Sex'] == 'M']

females_athlete = df_athlete_events.loc[df_athlete_events['Sex'] == 'F']

print('number of male athletes : ', males_athlete.Sex.count())

print('number of female athletes : ', females_athlete.Sex.count())
```

```python
df_medal_per_country.reset_index(inplace=True)
```

```python
#checking top countries in amount of medals
df_top = df_medal_per_country.nlargest(27, 'medal_count')
```

```python
plt.figure(figsize=(15,7))


sns.barplot(data=df_top,
            x='Team',
            y='medal_count',
           hue='Medal')
```
